### 6. Modelagem

#### 6.1 Machine Learning

```python
#Função: a= dataframe  e b='variável target'
def model(a,b):
    #Fatiando
    x = a.drop(b, axis=1)
    y = a[b]
    
    #Separação de Treino e Teste
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2)
    
    #Instanciamento
    randomForest = RandomForestClassifier(n_estimators=200)
    extraTree = ExtraTreesClassifier(n_estimators=200)
    logisticRegression = LogisticRegression(max_iter=2000)
    
    #Treino
    randomForest.fit(x_train, y_train)
    extraTree.fit(x_train, y_train)
    logisticRegression.fit(x_train, y_train)
    
    
    #Avaliação
    yrandomForest = randomForest.predict(x_test)
    yextraTree = extraTree.predict(x_test)
    ylogisticRegression = logisticRegression.predict(x_test)
    
    #Resultados
    previsao = x_test[['País',
                       'Ano',
                       'Local',
                       'Celular',
                       'Tamanho da Família',
                       'Idade',
                       'Gênero',
                       'Relação com Chefe de Família',
                       'Estado Civil',
                       'Nível Educacional',
                       'Tipo de Trabalho']]
    previsao[b] = y_test.values
    previsao['Random Forest'] = yrandomForest
    previsao['Extra Tree'] = yextraTree
    previsao['Logistic Regression'] = ylogisticRegression
    
    #Acurácia
    acuracia1 = randomForest.score(x_test, y_test)
    print(f'1) *Random Forest* acertou {(acuracia1 * 100).round(decimals=2)}% das previsões.')
    
    acuracia2 = extraTree.score(x_test, y_test)
    print(f'2) *Extra Tree* acertou {(acuracia2 * 100).round(decimals=2)}% das previsões.')
    
    acuracia3 = logisticRegression.score(x_test, y_test)
    print(f'3) *Logistic Regression* acertou {(acuracia3 * 100).round(decimals=2)}% das previsões.')
    
    
    print(previsao.shape)
    return previsao.head()
```

### 7. Acurácia dos Modelos de Machine Learning

```python
#Função: a= dataframe  e b='variável target'
model(df,'Conta Bancária')
```
